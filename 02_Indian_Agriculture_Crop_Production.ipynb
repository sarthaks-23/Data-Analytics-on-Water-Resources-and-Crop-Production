#### Dealing with various production units

```python
agri_df['Production Units'].unique()
```

```python
#agri_df[agri_df['Production Units']=='Nuts']['Crop'].unique()
```

```python
def std_unit(df):
    
    if df['Production Units'] == 'Nuts':
        new_production = df['Production']/50
        return new_production
    elif df['Production Units'] == 'Tonnes':
        return df['Production']
    else:
        new_production = df['Production']/4.5929637955183
        return new_production
```

```python
agri_df['New Production'] = agri_df.apply(std_unit,axis=1)
agri_df.sample(5)
```

```python
agri_df.drop(columns=['Production','Production Units'],inplace=True)
```
